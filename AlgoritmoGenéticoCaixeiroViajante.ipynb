```python
from deap import creator, base, tools, algorithms
```

```python
import numpy as np
```

```python
toolbox = base.Toolbox();  #base é uma classe e toolbox o objeto
n = 4
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("EstrIndividuo", list, fitness = creator.FitnessMin)

toolbox.register("Genes", np.random.permutation, n)
toolbox.register("Individuos", tools.initIterate, creator.EstrIndividuo, toolbox.Genes)
```

```python
#POPULAÇÃO
toolbox.register("Populacao", tools.initRepeat, list, toolbox.Individuos)
```

```python
pop = toolbox.Populacao(n = 10)
```

```python
dist = [[0, 7, 9, 2],
        [4, 0, 3, 7],
        [6, 7, 0, 8],
        [2, 3, 8, 0]]
```

```python
def aptidao(individuo):
  f = 0
  for i in range(n-1):
    local1 = individuo[i]
    local2 = individuo[i+1]
    f = f + dist[local1][local2]
  return f,
```

```python
toolbox.register("evaluate", aptidao)
```

```python
toolbox.register("select", tools.selTournament, tournsize = 2)
```

```python
toolbox.register("mate", tools.cxPartialyMatched)
```

```python
toolbox.register("mutate", tools.mutShuffleIndexes, indpb = 0.1)
```

```python
hof = tools.HallOfFame(1)
```

```python
result, log = algorithms.eaSimple(pop, toolbox, cxpb = 0.8, mutpb = 0.1, ngen = 30, halloffame = hof, verbose = True)
#NO TRABALHO VAMOS TER QUE TROCAR ESSE eaSimple
```
